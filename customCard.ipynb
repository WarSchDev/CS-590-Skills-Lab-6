```python
word_banks = {
    "birthday": {
        "greetings": ["Happy Birthday", "Wishing you a wonderful birthday", "Many happy returns of the day"],
        "adjectives": ["joyful", "magical", "spectacular", "unforgettable"],
        "wishes": ["May this year bring you everything you've dreamed of.", "Hope your day is filled with laughter and love.", "Wishing you a year as wonderful as you are."],
        "closings": ["With love", "Warmly", "Cheers", "Celebrating you"]
    },
    "christmas": {
        "greetings": ["Merry Christmas", "Season's Greetings", "Wishing you a very Merry Christmas"],
        "adjectives": ["warm", "magical", "joyful", "cozy", "festive"],
        "wishes": ["May your holidays be filled with warmth and cheer.", "Wishing you peace, joy, and plenty of hot cocoa.", "Hope your Christmas is merry and bright."],
        "closings": ["With holiday cheer", "Warmly", "With festive love", "Merry and bright"]
    },
    "halloween": {
        "greetings": ["Happy Halloween", "Wishing you a spooky Halloween", "Boo!", "Have a frightfully fun Halloween"],
        "adjectives": ["spooky", "eerie", "ghoulish", "wicked", "creepy", "bewitching"],
        "wishes": ["May your candy bag be overflowing!", "Hope your night is full of tricks and treats.", "May your costume be the scariest on the block."],
        "closings": ["Spookily yours", "Your friendly ghost", "Creepily yours", "BOO!"]
    },
    "valentine's day": {
        "greetings": ["Happy Valentine's Day", "Sending you love this Valentine's Day", "Be my Valentine", "With all my heart"],
        "adjectives": ["loving", "cherished", "adored", "wonderful", "sweet", "treasured"],
        "wishes": ["You make every day feel like Valentine's Day.", "May your day be as sweet as you are.", "Sending you all the love in the world today and always."],
        "closings": ["With love", "Forever yours", "All my love", "With a full heart"]
    },
    "death": {
        "greetings": ["With deepest condolences", "Thinking of you in this difficult time", "With heartfelt sympathy", "Sending you love and comfort"],
        "adjectives": ["cherished", "beloved", "irreplaceable", "deeply missed", "unforgettable"],
        "wishes": ["May you find peace and comfort in the memories you shared.", "Their memory will live on forever in the hearts of those who loved them.", "Please know that you are surrounded by love and support."],
        "closings": ["With deepest sympathy", "With love and support", "In loving memory", "Always here for you"]
    },
    "get well soon": {
        "greetings": ["Get Well Soon", "Sending healing thoughts your way", "Wishing you a speedy recovery", "Thinking of you"],
        "adjectives": ["strong", "resilient", "brave", "incredible", "tough", "remarkable"],
        "wishes": ["Wishing you a quick and smooth recovery.", "Hope you're back on your feet and feeling like yourself very soon.", "You are stronger than you know, and we're rooting for you."],
        "closings": ["Wishing you well", "With healing thoughts", "Rooting for you", "With care"]
    }
}
```

```python
import random

def generate_card(name, occasion, age=None):
    occasion = occasion.lower().strip()
    
    if occasion not in word_banks:
        print(f"Sorry, I don't have cards for '{occasion}' yet.")
        print(f"Available occasions: {', '.join(word_banks.keys())}")
        return
    
    #Build Card
    bank = word_banks[occasion]
    greeting  = random.choice(bank["greetings"])
    adjective = random.choice(bank["adjectives"])
    wish      = random.choice(bank["wishes"])
    closing   = random.choice(bank["closings"])

    if occasion == "birthday" and age is not None:
        body = f"Today you turn {age}, and what a {adjective} milestone it is! {wish}"
    elif occasion == "birthday" and age is None:
        body = f"What a {adjective} birthday this is! {wish}"
    elif occasion == "death":
        body = f"The loss you are facing is {adjective}, and words can hardly express how much we care. {wish}"
    else:
        body = f"What a {adjective} occasion this is. {wish}"
    
    # Print card
    print("=" * 50)
    print(f"{greeting}, {name}!")
    print()
    print(body)
    print()
    print(f"{closing},")
    print("Your friends")
    print("=" * 50)

def main():
    print("Welcome to the Greeting Card Generator!")
    print(f"Occasions available: {', '.join(word_banks.keys())}")
    print("Tip: You can also just enter a number and it will automatically make a birthday card!")
    print()

    name  = input("Enter the persons's name: ")
    entry = input("Enter the occasion: ").strip()

    age = None

    if entry.isdigit():
        occasion = "birthday"
        age = int(entry)
    elif entry.lower() == "birthday":
        occasion = "birthday"
        age_input = input("Enter their age (or press Enter to skip): ").strip()
        if age_input.isdigit():
            age = int(age_input)
    else:
        occasion = entry

    print()
    generate_card(name, occasion, age)

main()
```
